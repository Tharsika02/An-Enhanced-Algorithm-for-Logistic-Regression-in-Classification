```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
# Example using a CSV file
import pandas as pd

file_path = '/content/drive/My Drive/mammography_csv.csv'
df = pd.read_csv(file_path)

# Show the first 5 rows
print(df.head())


# Compute the correlation matrix
correlation_matrix = df.corr()

# Set the figure size
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Replace 'target' with the actual name of your class label column
class_counts = df['class'].value_counts()

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'], startangle=90)
plt.title('Binary Class Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
